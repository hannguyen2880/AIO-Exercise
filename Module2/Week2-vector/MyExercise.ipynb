# **1. Vector and matrix operations**

```python
import numpy as np
```

```python
def computeVectorLength(vector):
    len_of_vector = np.linalg.norm(vector)
    return len_of_vector
```

```python
vector = np.array([-2, 4, 9, 21])
result = computeVectorLength([vector])
print(round(result, 2))
```

```python
def computeDotProduct(vector1, vector2):
    result = np.dot(vector1, vector2)
    return result
```

```python
def matrix_multi_vector(matrix, vector):
    result = np.dot(matrix, vector)
    return result
```

```python
def matrix_multi_matrix(matrix1, matrix2):
    result = np.dot(matrix1, matrix2)
    return result
```

```python
def inverse_matrix(matrix):
    result = np.linalg.inv(matrix)
    return result
```

# **2. Eigenvector v√† eigenvalues**

```python
def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues,eigenvectors
```

```python
matrix = np.array([[0.9, 0.2], [0.1, 0.8]])
eigenvalues, eigenvectors = compute_eigenvalues_eigenvectors(matrix)
print(eigenvectors)
```

# **3. Cosine Similarity**

```python
def compute_cosine(v1, v2):
    cos_sim = computeDotProduct(v1, v2) / (computeVectorLength(v1)*computeVectorLength(v2))
    return cos_sim
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 0, 3, 0])
result = compute_cosine(x,y)
print(round(result, 3))
```

# **4. Background subtraction**

```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
```

```python
bg1_image = cv2.imread('/content/GreenBackground.png', 1)
bg1_image = cv2.resize(bg1_image, (678, 381))

ob_image = cv2.imread('/content/Object.png', 1)
ob_image = cv2.resize(ob_image, (678, 381))

bg2_image = cv2.imread('/content/NewBackground.jpg', 1)
bg2_image = cv2.resize(bg2_image, (678, 381))
```

```python
def computeDifference(bg_img, input_img):
    difference_three_channel = cv2.absdiff(bg_img, input_img) # Find the object in the image
    difference_single_channel = np.sum(difference_three_channel, axis=2) / 3.0 #Convert 3D image into 2D image (to process ob binary image)
    difference_single_channel = difference_single_channel.astype('uint8')
    return difference_single_channel
```

```python
difference_single_channel = computeDifference(bg1_image, ob_image)
cv2_imshow(difference_single_channel)
```

Now, we write a function to convert the image into binary image, and convert it into 3D again.

```python
def computeBinaryMask(difference_single_channel):
    difference_binary = np.where(difference_single_channel >= 15, 255, 0) #Convert into binary image
    #Convert the 2D image back to 3D to the same size as the background image to be combined
    difference_binary = np.stack((difference_binary,)*3, axis=-1)
    return difference_binary
```

```python
binary_mask = computeBinaryMask(difference_single_channel)
cv2_imshow(binary_mask)
```

**Combine**

```python
def replaceBackGround(bg1_image, bg2_image, ob_image):
    difference_single_channel = computeDifference(bg1_image,ob_image)
    binary_mask = computeBinaryMask(difference_single_channel)

    output = np.where(binary_mask==255, ob_image, bg2_image)

    return output
```

```python
output = replaceBackGround(bg1_image, bg2_image, ob_image)
cv2_imshow(output)
```
